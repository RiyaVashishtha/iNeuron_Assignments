# Q11. How do you group data in a Pandas DataFrame by a specific column and apply an aggregation function?

```python
grouped_df = df.groupby('PRODUCTLINE').agg({'SALES': 'sum'})
print(grouped_df)
```

```python
grouped_df1 = df.groupby('COUNTRY').agg({'PRODUCTLINE': 'count', 'SALES': 'sum'})
print(grouped_df1)
```

# Q12. How do you pivot a Pandas DataFrame?

```python
import pandas as pd

# Create a sample DataFrame
data = {
    'City': ['New York', 'London', 'Tokyo', 'London', 'Tokyo'],
    'Year': [2020, 2020, 2020, 2021, 2021],
    'Sales': [100, 200, 150, 250, 180]
}
df = pd.DataFrame(data)

# Pivot the DataFrame
pivot_df = df.pivot(index='Year', columns='City', values='Sales')

```

```python
print(pivot_df)
```

# Q25. What is the difference between .loc and .iloc in Pandas?

.loc and .iloc are used in slicing data.
.loc method includes the last element of the range passed in it, unlike iloc(). 
.loc() can accept the boolean data unlike iloc().

```python
df.loc[2:5]
```

```python
df.iloc[2:5]
```

```python
# selecting rows from 1 to 4 and columns from 2 to 4
# if we have to print row and column
df.iloc[1: 5, 2: 5]
```

```python
df.loc[1: 5]
```
